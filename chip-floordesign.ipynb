```python
import pandas as pd

# Replace with your actual path to the log file
log_path = '/content/drive/MyDrive/maskplace/maskplace/logs/log_adaptec1_2024-11-18-03-16-07_seed_42_pnm_128.csv'

# Load the log file with no header and define custom column names
data = pd.read_csv(log_path, header=None)
data.columns = ['Epoch', 'Raw Score', 'Moving Average Score', 'Other Metric']  # Replace with appropriate names
print(data.head())  # Preview the first few rows
```

```python
import matplotlib.pyplot as plt

# Plot Raw Score vs. Epoch
plt.figure(figsize=(10, 6))
plt.plot(data['Epoch'], data['Raw Score'], label='Raw Score')
plt.plot(data['Epoch'], data['Moving Average Score'], label='Moving Average Score', linestyle='--')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.title('Training Scores Over Epochs')
plt.legend()
plt.show()
```
